```python
# 모듈 설치
import pandas as pd
import matplotlib.pyplot as plt
import platform

# 파일 불러오기
df= pd.read_csv("./data/subwaytime.csv")

# 정보 확인
print(df.info())
print(df.head())
df
```

```python
df['호선명'].unique()
```

```python
df['07:00~08:59']=df.iloc[1:,11]+df.iloc[1:,13]
df
```

```python
df.columns[11]
```

```python
df_out_time=df[[df.columns[11],df.columns[13]]]
df_out_time.drop(0,inplace=True)
df_out_time = df_out_time.astype(int)
df_out_time['people']= df_out_time['Unnamed: 11']+ df_out_time['Unnamed: 13']
df_out_time.drop(['Unnamed: 11','Unnamed: 13'],inplace=True,axis=1)
df_out_time.insert(0,'호선명',df['호선명'])
df_out_time.insert(0,'지하철역',df['지하철역'])
df_out_time.set_index('지하철역',inplace=True)
df_out_time
```

```python
station_list=[]
people_list=[]
label_list=[]

for i in range(1,8):    
    max_people= df_out_time[df_out_time['호선명']==str(i) + '호선']
    people_list.append(max_people['people'].max())
    station_list.append(max_people['people'].idxmax())
    
```

```python
type(people_list[0])
```

```python
for i in range(1,8):
    label_list.append(str(i)+"호선:"+station_list[i-1])
    print('출근 시간대 {0}호선 최대 하차역은 {1}역이고 하차인원은 {2:,}명 입니다.'.format(i,station_list[i-1],people_list[i-1],))
```

```python
if (platform.system()	==	'Windows'):
    plt.rc('font',	family='Malgun Gothic')
else:
    plt.rc('font',	family='AppleGothic')

plt.figure(dpi=100)
plt.title("출근 시간대 지하철 노선별 최대 하차 인원 및 하차역")
plt.bar(label_list,people_list)
plt.xticks(rotation=80)
plt.show()
```
