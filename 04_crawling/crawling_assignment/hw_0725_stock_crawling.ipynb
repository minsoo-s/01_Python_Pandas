### 시가 총액 10위까지의 기업 정보를 크롤링
     * 네이버 금융 웹사이트 
           https://finance.naver.com/sise/sise_market_sum.naver        
       
     * 크롤링 항목 7개 출력
          - 종목명, 종목코드, 현재가, 전일가, 시가, 고가, 저가   

     * 네이버 웹사이트의 한글 깨짐 문제
          requests 라이브러리 사용   

     * 왼쪽 “시가 총액” 메뉴에서 상위 10개 기업 정보 크롤링   
          시가 총액 url: https://finance.naver.com/sise/sise_market_sum.naver     
          
     * 각 회사별 URL 및 이름을 리스트에 저장   
          <a href=“/item/main.naver?code=00590”, class=“tiahref=“le”> 삼성전자 </a>
***
### 프로그램 동작
     * 코스피 상위 10개 기업의 URL 및 기업 이름을 리스트에 저장
     * 메뉴에서 선택한 기업의 세부 주식 정보를 화면에 출력
     * 사용자가 -1을 입력할 때 까지 계속 반복함        

### 실행 모습
     --------------------------------------
      [ 네이버 코스피 상위 10대 기업 목록 ]	
     -------------------------------------
      [ 1] 삼성전자
      [ 2] LG에너지솔루션
      [ 3] SK하이닉스
      [ 4] 삼성바이오로직스
      [ 5] 삼성전자우
      [ 6] NAVER
      [ 7] 현대차
      [ 8] LG화학
      [ 9] 삼성SDI
      [10] 기아
      주가를 검색할 기업의 번호를 입력하세요(-1:	종료):	1
      https://finance.naver.com/item/main.naver?code=005930
      종목명:	삼성전자
      종목코드:	005930
      현재가:	61,300
      전일가:	61,800
      시가:	61,800
      고가:	62,200
      저가:	61,200
***

```python
# 모듈 설치 
# requests 라이브러리 사용 : 한글 깨짐 방지

from bs4 import	BeautifulSoup
from selenium import webdriver
import requests
import time

wd	=	webdriver.Chrome('C:\MINSU\chromedriver')

# 10개 주식 코드 정리 (국내 주가 1~10위 기업)
stock_nums = ['005930','373220','000660','207940','005935','005380','035420','051910','006400','000270']

top_ten_list=[]
url_list=[]

# 10위 기업 랭킹 순으로 한 기업 주가 사이트 열기
for stock_num in stock_nums: 
    naver_stock_url = 'https://finance.naver.com/item/main.naver?code=%s' %stock_num
    wd.get(naver_stock_url)
    url_list.append(naver_stock_url)
    
    html	=	wd.page_source
    soup	=	BeautifulSoup(html,'html.parser')
    
    
    stock_list = soup.find('dl',{'class':'blind'}).find_all('dd') # soup.find_all을 하면 다른 dl태그까지 불러오기 때문에 for문 안에서 find 사용.
    
    #각 요소 텍스트 변환 후 슬라이싱
    stock_name = (stock_list[1].text).split()[1]
    stock_code = (stock_list[2].text).split()[1]
    stock_present = (stock_list[3].text).split()[1] 
    stock_yesterday = (stock_list[4].text).split()[1]
    stock_start = (stock_list[5].text).split()[1]
    stock_high = (stock_list[6].text).split()[1]
    stock_low = (stock_list[8].text).split()[1]

    top_ten_list.append([stock_name,stock_code,stock_present,stock_yesterday,stock_start,stock_high,stock_low])


# 출력 화면 --------------------------------------------------------------------------------------------------
print('*' * 50)
print('[ 네이버 코스피 상위 10대 기업 목록 ]')
print('*' * 50)
print('[ 1] 삼성전자')
print('[ 2] LG에너지솔루션')
print('[ 3] SK하이닉스')
print('[ 4] 삼성바이오로직스')
print('[ 5] 삼성전자우')
print('[ 6] NAVER')
print('[ 7] 현대차')
print('[ 8] LG화학')
print('[ 9] 삼성SDI')
print('[10] 기아')

# 입력 반복 , -1일 때 종료.
while True:
    choice = int(input('주가를 검색할 기업의 번호를 입력하세요(-1: 종료):'))
    
    if choice == -1: break
    else:
        print(f'주가를 검색할 기업의 번호를 입력하세요(-1: 종료): {choice}')
        print(url_list[choice-1],end='')
        print(f'''
종목명:	{top_ten_list[choice-1][0]}
종목코드: {top_ten_list[choice-1][1]}
현재가: {top_ten_list[choice-1][2]}
전일가: {top_ten_list[choice-1][3]}
시가: {top_ten_list[choice-1][4]}
고가: {top_ten_list[choice-1][5]}
저가: {top_ten_list[choice-1][6]}
''')
```
