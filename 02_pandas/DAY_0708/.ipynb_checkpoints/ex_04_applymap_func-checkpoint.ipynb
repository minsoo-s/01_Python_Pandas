```python
# [ 사용자 정의 함수 ]---------------------------------------
def addValue(x):
    print(f'[DEBUG]------------- \n{x}')
    return x+x

def addTwoValue(x,y,z):
    print(f'[DEBUG]------------- \n {x},{y},{z}')
    return x+y+z

# [ Series 객체 단위 ]---------------------------------------

def missingValueCheck(sr):
    print(f'[DEBUG]------------- \n{sr}')
    return sr.isnull()

def missingValueSum(sr):
    print(f'[DEBUG]------------- \n{sr}')
    return sr.isnull().sum()

#def 
#    print(f'[DEBUG]------------- \n{sr}')
```
