# [ apply(함수명, axis=방향/축) 실습 ]

- 원소/요소에 함수 적용하기

- 입력 : Series의 요소 또는 DataFrame의 Series 즉, 행 또는 열

```python
# [기본] axis = 0 --- Column마다 함수 적용.

df1.apply(np.sum)
```

```python
# [기본] axis = 1 --- Row마다 함수 적용.
df1.apply(np.sum, axis=1)
```

```python
# 사용자 정의 함수 
def addValue(x):
    print(f'[DEBUG]------------- \n{x}')
    return x+x

def addTwoValue(x,y,z):
    print(f'[DEBUG]------------- \n {x},{y},{z}')
    return x+y+z
```

```python
# DataFrmae.apply() ---> Series 단위 
df1.apply(addValue)
```

```python
df1.apply(addValue, axis=1)
```

```python
# 변수 2개 이상인 경우, 초과하는 변수만큼 줘야 함.
df1.apply(addTwoValue, y=2000, z=150)
```

```python
# Series.apply() ---> 원소 단위도 가능. 원소일 경우: axis 무의미
df1['Age'].apply(addValue)
```

```python
# [ 사용자 정의 함수 ]

def missingValueCheck(sr):
    print(f'[DEBUG]------------- \n{sr}')
    return sr.isnull()

def missingValueSum(sr):
    print(f'[DEBUG]------------- \n{sr}')
    return sr.isnull().sum()
```

```python
df1.apply(missingValueCheck)
```

```python
df1.apply(missingValueSum, axis=1)
```

```python
df1.apply(missingValueSum)
```
