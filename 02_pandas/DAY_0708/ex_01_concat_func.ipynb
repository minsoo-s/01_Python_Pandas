## DF & SR 데이터들의 연결 실습

- 행(row) 방향 / 열 (column) 방향 데이터 연결하기

    - pandas.concat([data.data,...], axis=0, join = 'outer' )
        : 기본값 => 행방향, 합집합 ---> 없는 컬럼 데이터는 NaN으로 채우기.
    
    - pandas.DataFrame.append(DF)
        : 행방향만 연결 => 곧 사라질 메서드 (Deprecated)

```python
# 모듈 로딩 
import pandas as pd

# DF 2개 생성
df1 = pd.DataFrame({'A' : [1,2,3,4,5], 'B' : [11,22,33,44,55]})

df2 = pd.DataFrame({'B' : [1,2,3],'C' : [4,5,6]})
```

```python
df1
```

```python
df2
```

# (1) 2개 DF 연결 하기 [기본값]

```python
# [기본] 행방향, 모든 컬럼 합집합
allDF1 = pd.concat([df1,df2])
allDF1
```

```python
# [수정2] 행방향, 모든 컬럼 합집합 => 기존 인덱스 무시 ==> 새롭게 인덱스 설정
allDF2 = pd.concat([df1,df2] , ignore_index=True)
allDF2
```

```python
# [수정3] 행방향, 공통 컬럼 교집합 => 기존 인덱스 무시 ==> 새롭게 인덱스 설정
allDF3 = pd.concat([df1,df2] , ignore_index=True , join='inner')
allDF3
```

```python
# [수정4] 열 방향, 공통 행 교집합 
allDF4 = pd.concat([df1,df2] ,axis=1 , join='inner')
allDF4
```

```python
# [수정5] 열 방향, 모든 행 힙집합 
allDF5 = pd.concat([df1,df2] ,axis=1)
allDF5

# 이런 경우 B 2개 구분이 어려움. 
```

```python
# [수정6] 열 방향, 모든 행 힙집합 ,기존 걸럼명 무시 ==> 새롭게 컬럼명 지정
allDF6 = pd.concat([df1,df2] ,axis=1, ignore_index=True)
allDF6
```

```python
# [수정7] 열 방향, 공통 행 교집합, 기존 컬럼명 무시 ==> 새롭게 컬럼명 지정 
allDF7 = pd.concat([df1,df2] ,axis=1,join='inner', ignore_index=True)
allDF7
```
