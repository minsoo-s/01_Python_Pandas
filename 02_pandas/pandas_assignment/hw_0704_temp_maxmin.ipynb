- 2010년 1월~12월까지 최고온도-최저온도 데이터 추출

```python
# 모듈 로딩
import pandas as pd

# 파일 관련 함수 선언
DIR_PATH='../DATA/'
FILE_NAME=DIR_PATH+'weather.csv'
FILE_NAME
```

```python
# (1) CSV FILE => DataFrame으로 로딩
weatherDF=pd.read_csv(FILE_NAME)
weatherDF.info()
```

```python
weatherDF
```

```python
# 편리한 계산을 위해 필요없는 컬럼을 모두 지울 예정이므로 미리 month 컬럼의 데이터를 저장해둔다. 
# ---> month 컬럼은 최종결과물에서 월 표시용도
month=weatherDF['month']
```

```python
# weatherDF1 : 해당 달과 최댓값, 최소값만 구하면 되므로 필요없는 컬럼 삭제
weatherDF1=weatherDF.drop(['id','year','element','month'],axis=1)
weatherDF1
```

```python
# weatherDF2 : .max() 과 .min() 은 컬럼을 기준으로 계산되므로 행열을 전환한다. (1달 단위 최댓값,최솟값)
weatherDF2=weatherDF1.transpose()
weatherDF2
```

```python
# 짝수 열이 최댓값, 홀수 열이 최솟값 자료이므로 조건문 작성.
# weatherDF_month의 인덱스와 weatherDF2의 컬럼은 상응하므로 아래와 같이 작성.
print('[ 2010년 1월~12월까지 최고온도-최저온도 데이터 추출 ]')
print()
for i in weatherDF2.columns:
    if i % 2 ==0:
        print(f'{month[i]}월 최고 온도: {weatherDF2[i].max()}')
    else:       
        print(f'{month[i]}월 최저 온도: {weatherDF2[i].min()}')
        print('--------------------')
```
