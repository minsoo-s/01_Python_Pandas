# (3-3) 데이터 정규화

- 수치데이터 컬럼마다 값의 범위가 다름
- 값의 범위 0~ 1 / -1 ~ 1로 값을 설정

## (3-3-1) 방법 : 컬럼의 최대값의 절대값으로 전체 데이터 나누기

```python
auto_mpg['weight'].max(), auto_mpg['weight'].min(), abs(auto_mpg['weight'].max())
```

```python
auto_mpg['weight_nor'] = auto_mpg['weight'] / abs(auto_mpg['weight'].max())
auto_mpg['weight_nor'].min(), auto_mpg['weight_nor'].max()
```

## (3-3-2) 방법 : (컬럼의 데이터 - 최소값) /(최대값 - 최소값)

- 컬럼의 최대값 => max() 
- 컬럼의 최소값 => min()

```python
numerator = auto_mpg['weight'] - auto_mpg['weight'].min()
denominator = auto_mpg['weight'].max() - auto_mpg['weight'].min() 

auto_mpg['weight_nor_2'] = numerator/denominator
auto_mpg['weight_nor_2']
```

## (3-3-3) 방법 : 컬럼의 데이터 - 평균 / 표준편차 ==> Z-score

```python
numerator = auto_mpg['weight'] - auto_mpg['weight'].mean()
denominator = auto_mpg['weight'].std() 

auto_mpg['weight_nor_3'] = numerator/denominator
auto_mpg['weight_nor_3'].min(), auto_mpg['weight_nor_3'].max()
```

```python
auto_mpg1 = auto_mpg.drop(['car name','origin_label','origin','horsepower_label','weight_nor','weight_nor_2','weight_nor_3'],axis=1) 
auto_mpg1 
```

```python
def cal(col_name):
    auto_mpg1[col_name +'_nor'] = auto_mpg1[col_name] / abs(auto_mpg1[col_name].max())
    
    numerator = auto_mpg1[col_name] - auto_mpg1[col_name].min()
    denominator = auto_mpg1[col_name].max() - auto_mpg1[col_name].min() 
    auto_mpg1[col_name+'_nor_2'] = numerator/denominator
    
    numerator = auto_mpg1[col_name] - auto_mpg1[col_name].mean()
    denominator = auto_mpg1[col_name].std() 
    auto_mpg1[col_name+'_nor_3'] = numerator/denominator

for i in auto_mpg1.columns:
    cal(i)

auto_mpg1 = auto_mpg1.iloc[:,8:]
```
