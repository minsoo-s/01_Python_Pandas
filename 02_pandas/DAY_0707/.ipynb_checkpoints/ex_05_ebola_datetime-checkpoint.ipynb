```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수들
DIR='../Data/'
FILE= DIR + 'country_timeseries.csv'

FILE
```

```python
# (1) CSV => DataFrame 로딩
ebolaDF = pd.read_csv(FILE)
ebolaDF
```

```python
# (2) 데이터 확인
ebolaDF.info()
```

```python
# (3) 전처리
ebolaDF.isnull().sum() # 모두 NaN인 열은 없는 것으로 보임.
```

```python
# (3-1) NaN 데이터 처리 => 
ebolaDF=ebolaDF.dropna(how='all') # 행 전체가 모두 결측치일 경우 제거
ebolaDF=ebolaDF.fillna(0) # 모든 NaN을 0으로 치환
```

```python
# 결측치 처리 확인 : 삭제된 값 없음, 결측치 모두 치환됨.
ebolaDF.info()
```

```python
# (3-2) Data Type 정리

from datetime import date

# 오늘 날짜 칼럼 추가.
ebolaDF['today']= date.today()

# Date, today 컬럼 type 확인
print(type(ebolaDF['Date'].info()))
print(type(ebolaDF['today'].info()))
```

```python
# Date 컬럼을 datetime으로 변환.
ebolaDF['Date']= pd.to_datetime(ebolaDF['Date'])

# 오늘 날짜 datetime으로 변환.
ebolaDF['today']= pd.to_datetime(ebolaDF['today'])

# DATE 타입 변환 확인
print(ebolaDF['Date'])

# TODAY 타입변환 확인
print(ebolaDF['today'])
```

```python
# (4) 데이터 가공
# Ebola 바이러스 발병 후 몇일(Day) 지났는지에 대한 칼럼 추가.

# 몇 일 지났는지 계산 : 컬럼 마지막에 추가되어서 insert를 사용.
# ebolaDF['period']= ebolaDF['today'] - ebolaDF['Date']
# ebolaDF['period']

# 오늘로부터 경과 날짜 계산
ebolaDF.insert(0,'period',ebolaDF['today'] - ebolaDF['Date']) # 0번 컬럼에 추가
ebolaDF
```

```python
# 계산을 위해 넣었던 today 컬럼 삭제.
ebolaDF.drop("today", axis=1,inplace=True)
ebolaDF
```

```python
# Date를 인덱스로 삽입.
ebolaDF=ebolaDF.set_index('Date')
ebolaDF
```
