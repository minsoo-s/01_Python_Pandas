# [ Datetime => index로 설정 실습 ]

- set_index(컬럼) : 인덱스

```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수들 
DIR='../DATA/'
FILE = DIR + 'stock-data.csv'

FILE
```

```python
# (1) CSV FILE => DataFrame 로딩
stockDF=pd.read_csv(FILE)
stockDF.info()
```

```python
# (2) 데이터 확인
stockDF.head()
```

```python
stockDF.isnull().sum()
```

```python
# (3) 전처리 => NaN, Duplicate, DataType
# Date 컬럼 Object => datetime64

# 방법(1) series.astype()
stockDF['Date']= stockDF['Date'].astype('datetime64')
stockDF.info()
```

```python
# 방법(2) pandas.to_datatime()
stockDF['Date'] = pd.to_datetime(stockDF['Date'])
stockDF.info()
```

```python
stockDF.head()
```

#  [ 특정 데이터 추출 ] 

- Q1) 7월 데이터만 추출.
- Q2) 6월 20일부터 25일까지 데이터만 추출.

```python
stockDF_month07 = stockDF[stockDF['Date'].dt.month==7]
stockDF_month07
```

```python
stockDF[(stockDF['Date'].dt.month==6) & (stockDF['Date'].dt.day >=20)& (stockDF['Date'].dt.day <= 25)]
```

```python
day= stockDF['Date'].dt.day
month= stockDF['Date'].dt.month
cond= (day>=20) & (day<=25) & (month==6)
stockDF[cond]
```

```python
# index 변경 => Date 컬럼을 인덱스 설정
stockDF.set_index('Date', inplace=True)
stockDF
```

```python
stockDF.info()
```

```python
stockDF.index
```

# [ Q. datatime type 인덱스에서 2018년만 뽑기 ]

```python
stockDF2=pd.read_csv(FILE)
```

```python
stockDF2['Date']= stockDF2['Date'].astype('datetime64')
stockDF2.info()
```

```python
stockDF2.set_index('Date', inplace=True)
```

```python
stockDF2.index
```

```python
# 방법1
stockDF2.loc['2018']
```

```python
# 방법2
year = stockDF2.index.year
cond = (year == 2018)
stockDF2.loc[cond]
```

```python
# 방법3
stockDF2.loc[stockDF2.index.year == 2018]
```
