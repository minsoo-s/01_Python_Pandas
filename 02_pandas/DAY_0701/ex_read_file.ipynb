```python
# 3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 컬럼을 출력해 주세요.
result_setosa=irisDF[irisDF['Name']=='Iris-setosa']
result_versicolor=irisDF[irisDF['Name']=='Iris-versicolor']
result_virginica=irisDF[irisDF['Name']=='Iris-virginica']

print(f'Iris-setosa 최대:\n{result_setosa.max()}')
print(f'Iris-setosa 최소:\n{result_setosa.min()}')
print(f'Iris-versicolor 최대:\n{result_versicolor.max()}')
print(f'Iris-versicolor 최소:\n{result_versicolor.min()}')
print(f'Iris-virginica 최대:\n{result_virginica.max()}')
print(f'Iris-virginica 최소:\n{result_virginica.min()}')
```
