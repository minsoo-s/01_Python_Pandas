# DF  객체 연결------------------------------------------------

- 여러 개의 file에서 읽은 DF객체

- 메서드 실행 결과 DF객체

- 여러개의 DF객체를 연결 시켜서 하나의 객체 생성

- .concat( 속성 ->> 기본값은 합집합, join=inner : 열의 교집합)

```python
import pandas as pd
DIR_PATH='../DATA/'
FILE_NAME1=DIR_PATH+'concat_1.csv'
FILE_NAME2=DIR_PATH+'concat_2.csv'
FILE_NAME3=DIR_PATH+'concat_3.csv'

csvDF1=pd.read_csv(FILE_NAME1)
csvDF2=pd.read_csv(FILE_NAME2)
csvDF3=pd.read_csv(FILE_NAME3)
```

```python
print(csvDF1.info())
print(csvDF2.info())
print(csvDF3.info())
```

```python
# DF객체 연결하기(1) => pandas.concat([df1,df2...])
# 행 방향 연결, 기본 인덱스 유지, 컬럼 합집합 => 속성 : join='outer'
df= pd.concat([csvDF1,csvDF2,csvDF3])
df
```

```python
# DF객체 연결하기(2) => pandas.concat([df1,df2...])
# 행 방향 연결, 기본 인덱스 유지, 컬럼 교집합 => 속성 : join='inner'
df2= pd.concat([csvDF1,csvDF2,csvDF3],join='inner')
df2
```

```python
# DF객체 연결하기(3) => pandas.concat([df1,df2...])
# 행 방향 연결, 기본 인덱스 유지X 컬럼 교집합 => 속성 : join='inner'
df3= pd.concat([csvDF1,csvDF2,csvDF3],join='inner',ignore_index=True)
df3
```

```python
df3.info(),df3.index
```

```python
# DF 객체를 열방향으로 연결 => concat()
df4= pd.concat([csvDF1,csvDF2,csvDF3], axis=1)
df4               
```
