```python
# 모듈 로딩 -----------------
import pandas as pd

# 파일관련 변수 선언 -----------
DIR_PATH = '../Data/'
FILE_NAME = DIR_PATH + 'weather.csv'

FILE_NAME
```

```python
# (1) CSV File => DataFrame으로 로딩 -------
weatherDF = pd.read_csv(FILE_NAME)
weatherDF.info()
```

```python
# 과제 0704 - (1) 2010년 1월 ~ 12월 최고온도 - 최저온도 데이터 추출
col = ['id', 'year', 'month', 'element']

for i in range(int(len(weatherDF.index)/2)):
    print(f'{weatherDF["month"][2*i]}월의 최대값: {weatherDF.iloc[2*i].dropna().drop(col).max()}')
    print(f'{weatherDF["month"][2*i]}월의 최소값: {weatherDF.iloc[2*i+1].dropna().drop(col).min()}')    
```

```python
# 과제 0704 - (2) 서울에서 다른지역으로 이동한 년도별 인구수 / 데이터 추출 및 분석
import numpy as np


filename = DIR_PATH + '시도별 전출입 인구수.xlsx'
excelDF = pd.read_excel(filename)
```
