```python
# 모듈 로딩
import pandas as pd, random

# [ 데이터 생성 ]
# 이름, 생년월일, 나이 3개 컬럼으로 구성 
# 10명
# 생년월일 : 동년해, 동일날짜
# 이름 : 영문
df=pd.DataFrame({'NAME':['Tom kim','Jake Lee','John Park','A B','C D','E F','G I','J K','L M','N O'],
                   'BIRTH':pd.date_range('1994-01-07',periods=10, freq='M'),
                   'AGE':[random.randint(17,25) for count in range(10)]})
df
```

# [ str 데이터로 구성된 컬럼(Series)의 str 접근자 ]----------

- 전체 컬럼의 요소에 str 관련 메서드 사용 가능하도록 하는 접근자
- 사용법
        - 객체변수.[컬럼명].str.메서드() => upper, lower, split, replace...
        - 객체변수.[컬럼명].str.get(인덱스) => 분리 문자리스트에 특정 요소 추출

```python
# 문자열 분리

NAME_LIST = df['NAME'].str.split()
```

```python
#  분리된 문자열 리스트에서 특정값 가져오기

df['LAST_NAME'] = NAME_LIST.str.get(0)
df['FIRST_NAME'] = NAME_LIST.str.get(1)
df
```

```python
# str.split() : expand -> 나눠준 값을 시리즈로 하나씩 담음 -> 데이터프레임이 됨.
value = df['NAME'].str.split(expand=True)
type(value), value
```

```python
# 합치기, 컬럼명 변경
df2=pd.concat([df,value], axis=1)
df2.rename(columns={0:'Fname',1:'Lname'},inplace=True)
df2
```

# 실습 -------------------------------------------------------

- stock-data.csv

- 접근자 사용하여 Data 칼럼 분할 (Y, M, D)

- D일 기준으로 정렬

```python
# 1. 파일 불러오기
stockDF=pd.read_csv('../DATA/stock-data.csv')
stockDF
```

```python
# date 칼럼을 Y,M,D로 나누기
date = stockDF['Date'].str.split('-',expand=True)
```

```python
# 두 개 데이터프레임 합치기
stockDF2=pd.concat([stockDF,date],axis=1)

# 칼럼명 바꾸기
stockDF2.rename(columns={0:'year',1:'month',2:'day'},inplace=True)
stockDF2.
```

```python
# day를 기준으로 정렬하기
stockDF2.sort_values(by='day',inplace=True)
```
